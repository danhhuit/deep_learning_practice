1.   NUMPY VÀ PANDAS



```python
# Bài tập 3: Sinh viên tìm hiểu và viết chương trình sử dụng các hàm sau numpy.greater,
# numpy.greater_equal, numpy.less, và numpy.less_equal để so sánh 2 mảng 1 chiều, 2 chiều.
# Khởi tạo hai mảng 1 chiều
arr1_1d = np.array([10, 20, 30, 40, 50])
arr2_1d = np.array([5, 20, 35, 40, 55])

print("Mảng 1 (1D):")
print(arr1_1d)
print("Mảng 2 (1D):")
print(arr2_1d)
print("\n")

# numpy.greater(arr1, arr2): Trả về True nếu arr1 > arr2 theo từng phần tử
print("Kết quả numpy.greater(arr1_1d, arr2_1d) (arr1 > arr2):")
print(np.greater(arr1_1d, arr2_1d))

# numpy.greater_equal(arr1, arr2): Trả về True nếu arr1 >= arr2 theo từng phần tử
print("\nKết quả numpy.greater_equal(arr1_1d, arr2_1d) (arr1 >= arr2):")
print(np.greater_equal(arr1_1d, arr2_1d))

# numpy.less(arr1, arr2): Trả về True nếu arr1 < arr2 theo từng phần tử
print("\nKết quả numpy.less(arr1_1d, arr2_1d) (arr1 < arr2):")
print(np.less(arr1_1d, arr2_1d))

# numpy.less_equal(arr1, arr2): Trả về True nếu arr1 <= arr2 theo từng phần tử
print("\nKết quả numpy.less_equal(arr1_1d, arr2_1d) (arr1 <= arr2):")
print(np.less_equal(arr1_1d, arr2_1d))
```

```python
# Khởi tạo hai mảng 2 chiều
arr1_2d = np.array([[1, 2, 3], [4, 5, 6]])
arr2_2d = np.array([[0, 2, 4], [3, 5, 7]])

print("Mảng 1 (2D):")
print(arr1_2d)
print("Mảng 2 (2D):")
print(arr2_2d)
print("\n")

# numpy.greater(arr1, arr2)
print("Kết quả numpy.greater(arr1_2d, arr2_2d) (arr1 > arr2):")
print(np.greater(arr1_2d, arr2_2d))

# numpy.greater_equal(arr1, arr2)
print("\nKết quả numpy.greater_equal(arr1_2d, arr2_2d) (arr1 >= arr2):")
print(np.greater_equal(arr1_2d, arr2_2d))

# numpy.less(arr1, arr2)
print("\nKết quả numpy.less(arr1_2d, arr2_2d) (arr1 < arr2):")
print(np.less(arr1_2d, arr2_2d))

# numpy.less_equal(arr1, arr2)
print("\nKết quả numpy.less_equal(arr1_2d, arr2_2d) (arr1 <= arr2):")
print(np.less_equal(arr1_2d, arr2_2d))
```


2.   TENSORFLOW



```python
# Bài tập 13: Xây dựng model để tính tổng hai tensor x, y
# Khởi tạo hai tensor
x = tf.constant([1, 2, 3], dtype=tf.float32)
y = tf.constant([4, 5, 6], dtype=tf.float32)

print("Tensor x:", x.numpy())
print("Tensor y:", y.numpy())

# Tính tổng hai tensor
sum_tensor = tf.add(x, y)

print("\nTổng của hai tensor x và y:", sum_tensor.numpy())
```

```python
# Bài tập 14: Xây dựng model để tính hiệu và tích hai tensor x, y
# Khởi tạo hai tensor
x = tf.constant([[1, 2], [3, 4]], dtype=tf.float32)
y = tf.constant([[5, 6], [7, 8]], dtype=tf.float32)

print("Tensor x:\n", x.numpy())
print("Tensor y:\n", y.numpy())

# Tính hiệu hai tensor
difference_tensor = tf.subtract(x, y)

print("\nHiệu của hai tensor x và y:\n", difference_tensor.numpy())

# Tính tích hai tensor (element-wise multiplication)
product_tensor_elementwise = tf.multiply(x, y)

print("\nTích theo từng phần tử của hai tensor x và y:\n", product_tensor_elementwise.numpy())

# Tính tích ma trận (matrix multiplication) nếu các tensor là ma trận
# Đảm bảo điều kiện kích thước: (m, n) @ (n, p) -> (m, p)
product_tensor_matmul = tf.matmul(x, y)

print("\nTích ma trận của hai tensor x và y:\n", product_tensor_matmul.numpy())
```
