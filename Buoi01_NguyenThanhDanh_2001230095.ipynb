```python
# khai bao thu vien
import numpy as np
from numpy._core.numeric import ones

import pandas as pd
```

1. Numpy:

```python
# a. Tao mang 1 chieu:
single_list = [14, 7 , 2005, 6, 5 , 2026]
print('Danh sach cac phan tu: ', single_list)

# list --> numpy arr
single_arr = np.array(single_list)
print('Mang 1 chieu cac phan tu: ', single_arr)

# Cach khac:
single_arr2 = np.array([14, 7 , 2005, 6, 5 , 2026])
print('Mang 1 chieu cac phan tu: ', single_arr2)
```

```python
# b. Tao mang 2 chieu:
double_arr = np.array([
                        [1, 2, 3],
                        [4, 5, 6]
                      ])
print('Mang 2 chieu cac phan tu: ', double_arr)
print('Kich thuoc mang: ', double_arr.shape)
print('Kieu du lieu: ', double_arr.dtype)
```

```python
# c. Tao mang 3 chieu:
triple_arr = np.array([
                        [[1, 2, 3],
                         [4, 5, 6]],
                        [['hi', 8, 9],
                         [10, 11, 12]]
                      ])
print('Mang 3 chieu cac phan tu: ', triple_arr)
print('Kich thuoc mang: ', triple_arr.shape)
print('Kieu du lieu: ', triple_arr.dtype)
```

```python
# d. Tao Array voi gia tri ban dau la 0 hoac 1
zeros_matrix = np.zeros((2, 3)) # n dong, m cot
print('Mang 0: ', zeros_matrix)
print('Kich thuoc mang 0: ', zeros_matrix.shape)
print('Kieu du lieu mang 0: ', zeros_matrix.dtype)

ones_matrix = np.ones((2, 2))
print('Mang 1: ', ones_matrix)
print('Kich thuoc mang 1: ', ones_matrix.shape)
print('Kieu du lieu mang 1: ', ones_matrix.dtype)
```
